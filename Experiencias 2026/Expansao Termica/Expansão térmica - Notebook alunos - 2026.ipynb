## Introdução à Expansão Térmica dos Materiais

A expansão térmica é um fenómeno físico que descreve a variação das dimensões de um material com a temperatura. A origem microscópica deste efeito reside na interação entre os átomos ou moléculas que compõem a estrutura do material. À medida que a temperatura aumenta, a energia térmica das partículas cresce, resultando numa maior amplitude das vibrações atómicas. Devido à forma assimétrica do potencial de interação interatómico, esta vibração média desloca os átomos para posições ligeiramente mais afastadas, conduzindo à dilatação térmica macroscópica (ver <b>Figura 1 (a)</b>) . Este comportamento é quantificado pelo coeficiente de expansão térmica linear $\alpha_L$, definido como:

$$ \alpha_L = \dfrac{1}{L_0} \left( \dfrac{dL}{dT} \right)$$

onde $L_0$ representa o comprimento inicial do material e $\dfrac{dL}{dT}$ a taxa de variação do comprimento com a temperatura (ver <b>Figura 1(b)</b>). Materiais isotrópicos seguem uma expansão volumétrica descrita pelo coeficiente de expansão térmica volumétrica $\alpha_V = 3 \alpha_L$.

Na maioria dos sólidos, a expansão térmica é positiva, significando que o material se dilata ao aquecer e contrai ao arrefecer. No entanto, existem materiais com expansão térmica negativa (NTE), onde as dimensões diminuem com o aumento da temperatura. Este comportamento é frequentemente atribuído a modos vibracionais cooperativos, como movimentos de rede em estruturas cristalinas específicas, que promovem contração ao invés de dilatação. A busca por materiais com expansão térmica nula ou negativa tem sido um desafio secular na ciência dos materiais (ver materiais <i>invar</i>), motivado por aplicações que exigem estabilidade dimensional extrema. Componentes eletrónicos, ópticos e aeroespaciais requerem materiais com baixa ou nula expansão térmica para evitar falhas estruturais causadas por variações térmicas. 

Materiais compósitos, nos quais a combinação de diferentes coeficientes de expansão térmica permite a engenharia de estruturas dimensionalmente estáveis, têm sido amplamente explorados. A investigação continua focada na descoberta de novos materiais com NTE e na otimização das suas propriedades para aplicações tecnológicas avançadas.

## Estruturas Cristalinas e Difração de Raios-X

A estrutura cristalina de um material refere-se à organização periódica dos seus átomos no espaço. Os átomos que compõem os sólidos cristalinos apresentam uma organização ordenada que é descrita por uma rede cristalina, cujas propriedades influenciam diretamente fenómenos físicos como a expansão térmica, propriedades mecânicas, elétricas e resposta magnética. Cada ponto da rede cristalina representa a posição de um átomo ou grupo de átomos, e a repetição periódica dessa estrutura define a simetria e os parâmetros da célula unitária do cristal (a unidade mais básica da estrutura que se repete “infinitamente” para compor o material em 3D).

A determinação da estrutura cristalina (que tipo de simetria, e quais as distânicas interatómicas) de um material é frequentemente realizada por <b>difração de raios-X (XRD)</b>, uma técnica fundamental na ciência dos materiais. A difração ocorre quando um feixe de raios-X interage com os planos atómicos de um cristal, resultando em padrões característicos de interferência. O princípio que rege este fenómeno é descrito pela <b>lei de Bragg</b>, expressa como:


$$ n \lambda = 2 d \sin \theta$$

onde:

- $n$ é um número inteiro (ordem de difração);

- $\lambda$ é o comprimento de onda dos raios-X incidentes;

- $d$ é a distância entre os planos cristalinos;

- $\theta$ é o ângulo de incidência dos raios-X.


A análise dos picos de difração permite determinar os parâmetros da célula unitária e identificar a presença de fases cristalinas específicas. No contexto da expansão térmica, a análise da difração de raios-X (padrões) a diferentes temperaturas possibilita a medição direta da variação dos parâmetros de rede, fornecendo informações quantitativas sobre a expansão térmica do material.

Entre outras funções, o estudo dos padrões de difração pode ser feito a partir do ajuste da curva de cada pico de difração a uma função do tipo Gaussiano (representada na Figura 3):

$$ G \left( 2 \theta \right) = I_{max} \cdot \exp \left[ - \ln(2) \cdot \left( \dfrac{2 \theta - 2 \theta_{0}}{\omega} \right)^2 \right]$$


 onde $I_{max} \equiv$ intensidade do máximo; $2\theta_{0} \equiv$ posição do pico de difração; $\omega \equiv$ metade da largura a meia altura.

```python
def curve_fitting(x , y , func):
    """
    Função que efetua o ajuste considerando o método dos mínimos quadrados
    
    # Variáveis de entrada
    x - abcissas dos dados experimentais
    y - ordenadas dos dados experimentais
    func - nome da função (curva) que pretendemos ajustar aos dados experimentais
    
    # Variáveis de saída
    parms - lista com os valores dos parâmetros da curva de ajuste
    u_parms - lista com os valores das incertezas dos parâmetros de ajuste
    """
    
    parms , pov = spo.curve_fit( func , x , y  , maxfev = int(1e6))
    u_parms = np.sqrt( np.diagonal(pov) )
    
    return parms , u_parms
```
