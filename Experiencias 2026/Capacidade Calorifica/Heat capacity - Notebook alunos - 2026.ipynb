## Introdução às Medidas de Calor Específico

O calor específico ($C_p$) é uma propriedade termodinâmica fundamental que quantifica a quantidade de calor necessária para alterar a temperatura de um material por uma unidade de grau. Fornece informações muito relevantes sobre a capacidade de armazenamento de energia de um material e desempenha um papel essencial na compreensão das transições de fase que os materiais exibem, onde mudanças abruptas na capacidade térmica frequentemente sinalizam transformações nos estados estrutural, eletrónico ou magnético do material. Estas medições são particularmente relevantes para o estudo de materiais magnetocalóricos, multiferróicos, <i>phase change materials</i>, supercondutores e outros sistemas nos quais as variações térmicas impulsionam mudanças de fase.  As medições de calor específico são indispensáveis para a identificação de temperaturas de transição de fase, caracterização do calor latente, e compreensão das variações de entropia associadas às transições de fase. Nos materiais magnetocalóricos, por exemplo, anomalias em $C_p$ próximo da temperatura de Curie (temperatura de transição magnética) ajudam a determinar a variação de entropia isotérmica e a variação de temperatura adiabática, um parâmetro-chave para a eficiência da refrigeração. Ao combinar dados de calor específico com análises de magnetização e estrutura, os investigadores podem obter uma compreensão abrangente da cinética das transições de fase e do comportamento crítico em materiais avançados.

O princípio fundamental passa por aquecer a amostra de modo controlado, de forma a fornecer uma quantidade de energia conhecida, medindo a variação de temperatura resultante $\left( \Delta (t) \right)$. Isto é alcançado aquecendo a amostra a uma potência constante $\left( P \left( t \right) = P_0 \right)$ durante um certo período de tempo, seguido do seu arrefecimento $\left( P \left( t \right)  = 0 \right)$, conforme ilustrado na <b>Figura 3(c)</b>. Devido às condições de alto vácuo que são estabelecidas no interior da câmara do PPMS, a condução térmica entre a amostra e a <i>puck</i> é dominada pelo fenómeno de condução através dos fios. Desta forma, a partir da Lei de Newton para as transferências de calor, podemos determinar o $C_p$ efetuando um ajuste à dependência temporal da variação da temperatura da amostra:


$$ C_{total} \dfrac{d\Delta T}{dt} = - K_w \cdot \Delta T (t) - P (t) \, \Rightarrow \, \Delta T (t) = T(t) - T_b = \begin{cases} \dfrac{P_0 \cdot \tau}{C_{total}} + \left( \Delta T(0) - \dfrac{P_0 \cdot \tau}{C_{total}}  \right) \cdot \exp \left(- \dfrac t  \tau\right)  & 0 \le t \le t_{switch}  \\ \, \\ \Delta T \left(t_{switch}\right) \cdot \exp\left(- \dfrac{ t - t_{switch} }{ \tau} \right) & t > t_{switch}  \end{cases}$$
        
onde $\tau = \dfrac{C_{total}}{K_w} \Leftrightarrow C_{total} = K_w \cdot \tau$


Na prática, a determinação do $C_p$ requer duas medições independentes:

- <b>Adenda:</b> Medição do $C_p \equiv C_{adenda}$ do sistema $\rm{puck} + \rm{grease}$

- <b>Amostra:</b> Medição do $C_p \equiv C_{total}$ do sistema $\rm{puck} + \rm{grease} + \rm{amostra}$

No final, a capacidade calorífica da amostra será dada por:

$$ C_{sample} = C_{total} - C_{adenda}$$

```python
def curve_fitting(x , y , func):
    """
    Função que efetua o ajuste considerando o método dos mínimos quadrados
    
    # Variáveis de entrada
    x - abcissas dos dados experimentais
    y - ordenadas dos dados experimentais
    func - nome da função (curva) que pretendemos ajustar aos dados experimentais
    
    # Variáveis de saída
    parms - lista com os valores dos parâmetros da curva de ajuste
    u_parms - lista com os valores das incertezas dos parâmetros de ajuste
    """
    
    parms , pov = spo.curve_fit( func , x , y  , maxfev = int(1e6))
    u_parms = np.sqrt( np.diagonal(pov) )
    
    return parms , u_parms
```

## Determinação da dependência de $C_{total}$ na temperatura

<div class="alert alert-block alert-warning">
    <b>Parte 4 - Obtenção da dependência da capacidade calorífica da amostra com a temperatura.</b>
    <p> O modelo térmico e o algoritmo de ajuste de dados que foi considerado nas <b>Partes 2 e 3</b> assume que a capacidade calorífica $(C_P)$ será aproximadamente constante na gama de temperaturas de cada ciclo de medição $T(t)$. No entanto, esta suposição nem sempre é verdade. Por exemplo, em amostras que transitem de fase na gama de temperaturas em medição - como no nosso caso - surge um pico na temperatura de transição $(T_t)$ que viola a suposição considerada. Desta forma, para transições de fase de primeira ordem, a amostra é aquecida e arrefecida em torno de $T_t$, obtendo-se a dependência de $C_P$ na temperatura diretamente a partir do perfil $T(t)$, recorrendo ao método de <b><i>Single-curve slope analysis</i></b>.
        
De acordo com este método, a dependência da capacidade calorífica na temperatura será dada pela equação:
        
$$ C_{total} (T) = \dfrac{ - K_w \left( T - T_b \right) + P(T) }{ S(T) } $$
        
onde $S(T) = \dfrac{dT}{dt} \equiv$ declive da curva para cada valor de temperatura. 
        
</p>   
</div>

Função de filtragem - médias móveis

```python
def medias_moveis( dados , r ):
    """
    Função que efetua a filtragem por médias móveis
    
    # Variáveis de entrada
    dados - array a ser filtrado
    r - vizinhos até onde se determina o valor médio
    
    # Variável de saída
    array filtrado
    """
    N = len(dados)
    lista=[]
    for i in range(N):
        Yk=0
        
        block_size = 2 * r + 1
        
        if i < r:
            block_size = r + 1 + i
            
        elif N - 1 - i < r:
            block_size = r + 1 + ( N - 1 - i )
        
        
        for j in range(-r,r+1):
            
            if (i+j) < 0:
                continue
            
            elif (i+j) >= N:
                continue
            
            Yk += dados[i+j] / block_size
            
        lista.append(Yk)
    return np.array(lista)
```

<div class="alert alert-block alert-warning">
    <p> <b>Se chegaste a esta parte estás de parabéns!</b> Agora está na altura de testar a tua independência e pensamento crítico.</p>
    <p> Neste ponto já deverás ter identificado a gama de temperaturas associada à transição de fase do material em estudo (hematite), confirmando a presença de uma <b>histerese</b> - diferença entre aquecimento e arrefecimento.</p> 
    <p> Na verdade, a partir da dependência na temperatura da capacidade calorífica em torno da transição, existe ainda uma propriedade física adicional que podes determinar: a <b>energia de transição de fase</b> $\left(\Delta E_t\right)$, ou seja a energia que é libertada pelo sistema durante a sua transição. Num contexto tecnológico este parâmetro é fundamental para determinar o valor máximo de energia que é possível extrair do sistema.</p>
    <p> Na prática, para determinar $\Delta E_t$, teremos de calcular a área da curva $C_{sample} (T)$ em torno da transição, ou seja, o seu <b>integral</b>, em aquecimento e arrefecimento.</p>
    <p> Para o efeito deverás proceder da seguinte forma: </p>
    <ol>
        <li>Determinar a temperatura da transição $\left(T_t\right)$; </li>
        <li>Estabelecer a largura da janela da transição de fase $\left( \delta T \right)$, ou seja, a gama de temperaturas durante a qual a transição ocorre; </li>
        <li>Calcular o integral da curva considerando esses parâmetros: </li>
    $$ \Delta E_t = \int_{T_t - \delta T/2}^{T_t + \delta T/2} C_{sample} (T') \, dT'$$</ol>
        <p><b>Nota:</b> Para calcular o integral poderá ser útil considerar o módulo do <i>scipy</i> "<b>scipy.integrate.simpson</b>"</p>
